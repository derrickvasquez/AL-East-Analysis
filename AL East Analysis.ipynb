```python
# Importing the necessary modules and packages
import pybaseball as pyball
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Generating AL East team data for the 2021 season 
year = 2021
yankees = pyball.schedule_and_record(year, 'NYY')
red_sox = pyball.schedule_and_record(year, 'BOS')
rays = pyball.schedule_and_record(year, 'TB')
orioles = pyball.schedule_and_record(year, 'BAL')
blue_jays = pyball.schedule_and_record(year, 'TOR')
```

```python
# Gets wins-to-date, losses-to-date, and win_percent-to-date for each team
yankees['Wins'] = np.where(yankees['W/L']=='W', 1, (np.where(yankees['W/L']=='W-wo', 1, 0))).cumsum()
yankees['Losses'] = np.where(yankees['W/L']=='L', 1, (np.where(yankees['W/L']=='L-wo', 1, 0))).cumsum()
yankees['Win_Percentage'] = yankees['Wins'] / (yankees['Wins'] + yankees['Losses'])

red_sox['Wins'] = np.where(red_sox['W/L']=='W', 1, (np.where(red_sox['W/L']=='W-wo', 1, 0))).cumsum()
red_sox['Losses'] = np.where(red_sox['W/L']=='L', 1, (np.where(red_sox['W/L']=='L-wo', 1, 0))).cumsum()
red_sox['Win_Percentage'] = red_sox['Wins'] / (red_sox['Wins'] + red_sox['Losses'])

rays['Wins'] = np.where(rays['W/L']=='W', 1, (np.where(rays['W/L']=='W-wo', 1, 0))).cumsum()
rays['Losses'] = np.where(rays['W/L']=='L', 1, (np.where(rays['W/L']=='L-wo', 1, 0))).cumsum()
rays['Win_Percentage'] = rays['Wins'] / (rays['Wins'] + rays['Losses'])

orioles['Wins'] = np.where(orioles['W/L']=='W', 1, (np.where(orioles['W/L']=='W-wo', 1, 0))).cumsum()
orioles['Losses'] = np.where(orioles['W/L']=='L', 1, (np.where(orioles['W/L']=='L-wo', 1, 0))).cumsum()
orioles['Win_Percentage'] = orioles['Wins'] / (orioles['Wins'] + orioles['Losses'])

blue_jays['Wins'] = np.where(blue_jays['W/L']=='W', 1, (np.where(blue_jays['W/L']=='W-wo', 1, 0))).cumsum()
blue_jays['Losses'] = np.where(blue_jays['W/L']=='L', 1, (np.where(blue_jays['W/L']=='L-wo', 1, 0))).cumsum()
blue_jays['Win_Percentage'] = blue_jays['Wins'] / (blue_jays['Wins'] + blue_jays['Losses'])
```

```python
# Graph for Total Wins for each team throughout the 2021 Season
plt.rcParams["figure.figsize"] = (8,6)

plt.plot(yankees['Wins'], label='Yankees', c='navy')
plt.plot(red_sox['Wins'], label='Red Sox', c='red')
plt.plot(rays['Wins'], label='Rays', c='green')
plt.plot(orioles['Wins'], label='Orioles', c='orange')
plt.plot(blue_jays['Wins'], label='Blue Jays', c='#348feb')

plt.xticks(np.arange(0, len(yankees.index), step=10))
plt.xlabel('Game')
plt.ylabel('Wins')

plt.legend(loc='lower right')

plt.title("AL East Wins Comparison ({})".format(year))
```

```python
# Graph for Win Percentage for each team throughout the 2021 Season
plt.rcParams["figure.figsize"] = (8,6)

plt.plot(yankees['Win_Percentage'], label='Yankees', c='navy')
plt.plot(red_sox['Win_Percentage'], label='Red Sox', c='red')
plt.plot(rays['Win_Percentage'], label='Rays', c='green')
plt.plot(orioles['Win_Percentage'], label='Orioles', c='orange')
plt.plot(blue_jays['Win_Percentage'], label='Blue Jays', c='#348feb')

plt.xticks(np.arange(0, len(yankees.index), step=10))
plt.xlabel('Game')
plt.ylabel('Win Percentage')

plt.legend(loc='lower right')

plt.title("AL East Win Percentage Comparison ({})".format(year))
```

```python
# Creating new columns for Runs and Runs Allowed from the 2021 season
yankees['Total_Runs'] = yankees['R'].cumsum()
yankees['Total_RA'] = yankees['RA'].cumsum()

red_sox['Total_Runs'] = red_sox['R'].cumsum()
red_sox['Total_RA'] = red_sox['RA'].cumsum()

rays['Total_Runs'] = rays['R'].cumsum()
rays['Total_RA'] = rays['RA'].cumsum()

orioles['Total_Runs'] = orioles['R'].cumsum()
orioles['Total_RA'] = orioles['RA'].cumsum()

blue_jays['Total_Runs'] = blue_jays['R'].cumsum()
blue_jays['Total_RA'] = blue_jays['RA'].cumsum()
```

```python
# Graph Runs and Runs Allowed Comparisons
plt.rcParams["figure.figsize"] = (16,8)

plt.subplot(1,2,1)
plt.plot(yankees['Total_Runs'], label='Yankees', c='navy')
plt.plot(red_sox['Total_Runs'], label='Red Sox', c='red')
plt.plot(rays['Total_Runs'], label='Rays', c='green')
plt.plot(orioles['Total_Runs'], label='Orioles', c='orange')
plt.plot(blue_jays['Total_Runs'], label='Blue Jays', c='#348feb')

plt.xticks(np.arange(0, len(yankees.index), step=10))
plt.xlabel('Game')
plt.ylabel('Runs')

plt.title("AL East Runs Comparison ({})".format(year))
plt.legend(loc='lower right')

plt.subplot(1,2,2)
plt.plot(yankees['Total_RA'], label='Yankees', c='navy')
plt.plot(red_sox['Total_RA'], label='Red Sox', c='red')
plt.plot(rays['Total_RA'], label='Rays', c='green')
plt.plot(orioles['Total_RA'], label='Orioles', c='orange')
plt.plot(blue_jays['Total_RA'], label='Blue Jays', c='#348feb')

plt.xticks(np.arange(0, len(yankees.index), step=10))
plt.xlabel('Game')
plt.ylabel('Runs Allowed')

plt.title("AL East Runs Allowed Comparison ({})".format(year))
plt.legend(loc='lower right')
```

```python
# Total Runs 2021 in tabular format
runs_scored_2021 = {'Team': ['Yankees', 'Red Sox', 'Rays', 'Orioles', 'Blue Jays'],
        'Total Runs': [yankees['R'].sum(), red_sox['R'].sum(), rays['R'].sum(), orioles['R'].sum(), blue_jays['R'].sum()]
        }

runs_scored_2021_df = pd.DataFrame(runs_scored_2021)

print(runs_scored_2021_df.sort_values(by='Total Runs', ascending=False))
```

```python
# Total Runs Allowed 2021 in tabular format
runs_allowed_2021 = {'Team': ['Yankees', 'Red Sox', 'Rays', 'Orioles', 'Blue Jays'],
        'Total Runs Allowed': [yankees['RA'].sum(), red_sox['RA'].sum(), rays['RA'].sum(), orioles['RA'].sum(), blue_jays['RA'].sum()]
        }

runs_allowed_2021_df = pd.DataFrame(runs_allowed_2021)

print(runs_allowed_2021_df.sort_values(by='Total Runs Allowed', ascending=False))
```

```python
# Creating new columns for Pitching Stats
from pybaseball import team_pitching
team_pitching_stats = team_pitching(2021, end_season=2021, league='AL', ind=0)
team_pitching_stats
```

```python
# Creating new dataframe for Pitching Stats
yankees_pitching = team_pitching_stats[team_pitching_stats['Team'] == 'NYY']
red_sox_pitching = team_pitching_stats[team_pitching_stats['Team'] == 'BOS']
rays_pitching = team_pitching_stats[team_pitching_stats['Team'] == 'TBR']
orioles_pitching = team_pitching_stats[team_pitching_stats['Team'] == 'BAL']
blue_jays_pitching = team_pitching_stats[team_pitching_stats['Team'] == 'TOR']

era_2021 = {'Team': ['Yankees', 'Red Sox', 'Rays', 'Orioles', 'Blue Jays'],
        'ERA': [yankees_pitching['ERA'].sum(), red_sox_pitching['ERA'].sum(), rays_pitching['ERA'].sum(), orioles_pitching['ERA'].sum(), blue_jays_pitching['ERA'].sum()]
        }

era_2021_df = pd.DataFrame(era_2021)

print(era_2021_df.sort_values(by='ERA', ascending=False))
```
